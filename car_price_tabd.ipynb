## Preprocessing

```python
df1.mark = pd.Categorical(df1.mark)
df1.mark = df1.mark.cat.codes

df1.Box = pd.Categorical(df1.Box)
df1.Box = df1.Box.cat.codes

df1.Drive_unit = pd.Categorical(df1.Drive_unit)
df1.Drive_unit = df1.Drive_unit.cat.codes

df1.Country = pd.Categorical(df1.Country)
df1.Country = df1.Country.cat.codes

df1.Engine = pd.Categorical(df1.Engine)
df1.Engine = df1.Engine.cat.codes

df1.Color = pd.Categorical(df1.Color)
df1.Color = df1.Color.cat.codes

df1.Car_type = pd.Categorical(df1.Car_type)
df1.Car_type = df1.Car_type.cat.codes

df1.Generation = pd.Categorical(df1.Generation)
df1.Generation = df1.Generation.cat.codes

df1.Model = pd.Categorical(df1.Model)
df1.Model = df1.Model.cat.codes
```

### Feature Selection

#### Method Univariate Selection

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

my_feature = 13

my_test = SelectKBest(chi2, k = my_feature)
my_fit = my_test.fit(X, Y)

np.set_printoptions(precision = 3)

scores =  my_fit.scores_
 
plt.figure(dpi=100)
plt.bar(range(len(feature_list)), scores)
plt.xticks(range(len(feature_list)), feature_list, rotation='vertical')
plt.show()    
    
    
X_new = my_fit.transform(X)

new_list = my_fit.get_support(indices=True)
print('Номера лучших фич: ', new_list)

print('Размер нового массива: ', X_new.shape)
print('------------------------')

my_dict = {}

for i in range(len(new_list)):
    j = new_list[i]
    my_dict.update({my_fit.scores_[j]: feature_list[j]}) 
    
print(my_dict)

from operator import itemgetter

dict(sorted(my_dict.items(), key = itemgetter(0), reverse=True))
```

#### Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

YY=Y.ravel()

model = LinearRegression()

rfe = RFE(model,  n_features_to_select = 13, step=1 )

fit = rfe.fit(x_scaled, YY)

print('Число важных фич: ', rfe.n_features_)
print('---------------------------------------')

print('"Ранг фич": ', rfe.ranking_)

for i in range(len(rfe.ranking_)):
    if rfe.ranking_[i] == 1:
        print(feature_list[i],' - ', rfe.ranking_[i])
```

#### RandomForestRegressor feature selection

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
rf = RandomForestRegressor()
rf.fit(x_scaled, YY)
print("Фичи, отсортированые по важности:")
print (sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), feature_list), 
             reverse=True))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
print('Size of the test collection: ', X_test.shape)
print('Size of the training collection: ', X_train.shape)
print()
```

```python
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import ShuffleSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

### K Nearest Neighbors

```python
knn_params = {
    'n_neighbors' : [3,5,10,15,20,25,35],
    'weights' : ['uniform', 'distance']
}

knn = KNeighborsClassifier(n_jobs=-1)
mycv = ShuffleSplit(n_splits=10, test_size=0.3, random_state=384)

knn_grid = GridSearchCV(estimator = knn, param_grid = knn_params, cv = mycv, scoring='f1_macro')
knn_grid.fit(X_train, y_train)
knn_grid_best_estimator = knn_grid.best_estimator_

predicted_knn_grid = knn_grid.predict(X_test)

print ('best estimator: \n', knn_grid_best_estimator)
print ('\nTrain F1: \n', knn_grid.score(X_train, y_train))
print ('\nTest F1: \n', knn_grid.score(X_test, y_test))

print ('\nbest_params: \n', knn_grid.best_params_)

print(metrics.classification_report(y_test, predicted_knn_grid))

print('confusion matrix')
print(metrics.confusion_matrix(y_test, predicted_knn_grid))

%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
cmap=plt.cm.Blues
ax.imshow(metrics.confusion_matrix(y_test, predicted_knn_grid), cmap=cmap)
fig.set_figwidth(6)     
fig.set_figheight(6)    
plt.show()
```

### The Random Forest Classifier

```python
import time

t1 = time.time()
param_rf = { 
    'n_estimators': [50],
    'max_depth' : np.linspace(1, 32, 32, endpoint=True),
    'criterion' : ['gini', 'entropy']
}


rf = RandomForestClassifier(n_jobs=-1, random_state=6763)
mycv = ShuffleSplit(n_splits=10, test_size=0.3, random_state=0)
rf = GridSearchCV(estimator = rf, param_grid=param_rf, cv= mycv, scoring='f1')

rf.fit(X, Y.ravel())

rf_best_estimator = rf.best_estimator_

predicted_rf = rf.predict(X_test)

print(rf.best_params_)
print(rf.best_score_)
t2 = time.time()
print('расчет закончен', t2-t1)

print ('best estimator: \n', rf_best_estimator)
print ('\nTrain F1: \n', rf.score(X_train, y_train))
print ('\nTest F1: \n', rf.score(X_test, y_test))

print ('\nbest_params: \n', rf.best_params_)

print(metrics.classification_report(y_test, predicted_rf))

print('confusion matrix')
print(metrics.confusion_matrix(y_test, predicted_rf))

%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
cmap=plt.cm.Blues
ax.imshow(metrics.confusion_matrix(y_test, predicted_rf), cmap=cmap)
fig.set_figwidth(6)     
fig.set_figheight(6)    
plt.show()
```
